```python
import numpy as np
import math
import matplotlib.pyplot as plt
import random
```

```python
def ArithmeticMean(arg_array):
    S  = 0
    
    for elem in arg_array:
        if type(elem) != int :
            return "Incorrect type"
        S += elem
        
    elems_num = len (arg_array)
    
    return S / elems_num
```

```python
def StandartDeviation(arg_array):
    
    arith_mean = ArithmeticMean(arg_array)
    print(arith_mean)
    S = 0 
    n = len (arg_array)
    for i in arg_array:
         S +=  ((i - arith_mean)**2)
    
    return math.sqrt(S /n)
        
        
        
     
 
    
```

```python
list_array = list(range(10))
print (StandartDeviation(list_array))
```

```python
def SrednayaLiniya(arg_array,interval):
    Y = []
    #a = np.random.randint(0,10,size=10)
    print(arg_array)
    for i in range (interval,len(arg_array)-interval):
            S= ArithmeticMean(arg_array[i-interval : i+1+interval])
        
            Y.append(S)    
    return Y
```

```python
def Online(arg_array , size, dop):
    otv = []
    
    first = dop+arg_array
    otv +=  SrednayaLiniya(first,size)
     
    return otv
```
